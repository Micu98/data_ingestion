# Web Scraping with Python and Beautifulsoup

## Libraries and settings

```python
import requests
import xlwt
from bs4 import BeautifulSoup
import os

# Get current working directory
print(os.getcwd())
```

## Create workbook

```python
wb = xlwt.Workbook()
ws = wb.add_sheet('Emails')
ws.write(0,0,'Emails')
```

## Initialize list and counter

```python
emailList = []
r = 0
```

## Add URL to search for emails

```python
# Add url of the page you want to scrape to urlString
# urlString = 'https://www.zh.ch/de/bildung/schulen.html'
urlString='https://www.agvs-upsa.ch/de/verband/mitgliederverzeichnis/liste'
```

## Function to extract emails and store them in the list

```python
def emailExtractor(urlString):
    getH=requests.get(urlString)
    h=getH.content
    soup=BeautifulSoup(h,'html.parser') # The HTML content (h) is parsed (analyze) using the 'html.parser' option, which allows BeautifulSoup to create a parse tree 
                                            #from the page's HTML structure, making it easier to extract specific elements.
    mailtos = soup.select('a[href^=mailto]')
    for i in mailtos: #mailto = which is the HTML convention for email links.
        href=i['href'] # href attribute in html doc
        try:
            str1, str2 = href.split(':')
        except ValueError:
            break
        
        emailList.append(str2)

emailExtractor(urlString)
```

```python
import requests
from bs4 import BeautifulSoup

def phoneExtractor(urlString):
    # Anfrage an die URL senden
    getH = requests.get(urlString)
    h = getH.content
    soup = BeautifulSoup(h, 'html.parser') 
    
    # Alle div-Tags mit Telefonnummern finden
    phone_divs = soup.select('div.views-field-field-telefon .field-content')
    
    phoneList = []  # Liste für Telefonnummern
    for div in phone_divs:
        phone_number = div.get_text(strip=True)  # Telefonnummer extrahieren und formatieren
        phoneList.append(phone_number)
    
    return phoneList

# Beispiel-Aufruf der Funktion
phone_numbers = phoneExtractor(urlString)
print(phone_numbers)
```

```python
# Parts of the function ...
getH = requests.get(urlString)
print(getH) # Die Zahl in den eckigen Klammern ist der Statuscode der Antwort (z. B., 200 für „Erfolg“, 404 für „Seite nicht gefunden“, 500 für einen Serverfehler).
```
